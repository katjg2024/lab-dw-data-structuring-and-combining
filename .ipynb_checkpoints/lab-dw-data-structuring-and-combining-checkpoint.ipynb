In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:

```python
import pandas as pd
dt = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv")
dt_1 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv")
dt_2 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv")
```

```python
dt_1.shape
```

```python
dt_2.shape
```

```python
4008 + 996 + 7070 - 3
```

```python
dt_2.nunique()
```

```python
dt.columns = dt.columns.str.lower().str.replace(" ", "_")
dt.rename(columns = {"st": "state"}, inplace=True)
dt

dt_1.columns = dt_1.columns.str.lower().str.replace(" ", "_")
dt_1.rename(columns = {"st": "state"}, inplace=True)
dt_1

dt_2.columns = dt_2.columns.str.lower().str.replace(" ", "_")
dt_2.rename(columns = {"st": "state"}, inplace=True)
dt_2
```

```python
dt['customer_lifetime_value'] = dt['customer_lifetime_value'].replace('%',"")
dt['customer_lifetime_value'] = dt['customer_lifetime_value'].astype(float) / 100

dt["customer_lifetime_value"]=pd.to_numeric(dt["customer_lifetime_value"])


dt_1['customer_lifetime_value'] = dt_1['customer_lifetime_value'].str.replace('%',"")
dt_1['customer_lifetime_value'] = dt_1['customer_lifetime_value'].astype(float) / 100

dt_1["customer_lifetime_value"]=pd.to_numeric(dt_1["customer_lifetime_value"])


dt_2['customer_lifetime_value'] = dt_2['customer_lifetime_value'].replace('%',"")
dt_2['customer_lifetime_value'] = dt_2['customer_lifetime_value'].astype(float) / 100

dt_2["customer_lifetime_value"]=pd.to_numeric(dt_2["customer_lifetime_value"])
dt_2
```

```python
dt['number_of_open_complaints'] = dt['number_of_open_complaints'].apply(lambda x: x.split("/")[1] if isinstance(x, str) else x)
dt['number_of_open_complaints'] = pd.to_numeric(dt['number_of_open_complaints'])

dt_1['number_of_open_complaints'] = dt_1['number_of_open_complaints'].apply(lambda x: x.split("/")[1] if isinstance(x, str) else x)
dt_1['number_of_open_complaints'] = pd.to_numeric(dt_1['number_of_open_complaints'])

dt_2['number_of_open_complaints'] = dt_2['number_of_open_complaints'].apply(lambda x: x.split("/")[1] if isinstance(x, str) else x)
dt_2['number_of_open_complaints'] = pd.to_numeric(dt_2['number_of_open_complaints'])
```

```python
merged_dt = pd.merge(dt, dt_1, on=['customer', 'state', 'gender', 'education', 'customer_lifetime_value',
       'income', 'monthly_premium_auto', 'number_of_open_complaints', 'policy_type', 'vehicle_class', 'total_claim_amount'], how='outer')
```

```python
merged_all = pd.merge(merged_dt, dt_2, on=['customer', 'state', 'gender', 'education', 'customer_lifetime_value',
       'income', 'monthly_premium_auto', 'number_of_open_complaints', 'policy_type', 'vehicle_class', 'total_claim_amount'], how='outer')
```

```python
merged_all
```

```python
merged_all['state'].value_counts()
```

```python
state_names = {'Cali': 'California', 'WA': 'Washington', 'AZ': 'Arizona'}
merged_all['state'].replace(state_names, inplace=True)
merged_all['state'].value_counts()
```

```python
edu_name = {"Bachelors": "Bachelor"}
merged_all.replace(edu_name, inplace=True)
merged_all['education'].value_counts()
```

```python
car_name = {"Sports Car": "Luxury", "Luxury SUV": "Luxury", "Luxury Car": "Luxury"}
merged_all["vehicle_class"].replace(car_name, inplace=True)
merged_all["vehicle_class"].value_counts()
```

```python
merged_all['number_of_open_complaints'].value_counts()
```

```python
merged_all.isna().any()
```

```python
merged_all.dropna(how='all', inplace=True) #dropping all rows with all NaN values
merged_all.isna().any()
```

```python
merged_all["gender"].value_counts()
```

```python
gender_map = {'Male': 'M', 'female': 'F', 'Femal': 'F'}
merged_all['gender'].replace(gender_map, inplace=True)
merged_all['gender'].value_counts()
```

```python
merged_all.gender.fillna("F", inplace = True)
merged_all.isna().any()
```

```python
merged_all.customer_lifetime_value.fillna(merged_all.customer_lifetime_value.mean(), inplace=True)
merged_all.isna().any()
```
